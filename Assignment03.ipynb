Python program for displaying statitical details like percentile, mean,standard deviation for the data set

```python
def Cal_Mean(df):
    x = df[['Experience_Years', 'Age', 'Salary']].mean()
    print(x)

def Cal_Perc(df):
    x = df[['Experience_Years', 'Age', 'Salary']].quantile()
    print(x)
    
def Cal_Median(df):
    x = df[['Experience_Years', 'Age', 'Salary']].median()
    print(x)

def Cal_Max(df):
    x = df[['Experience_Years', 'Age', 'Salary']].max()
    print(x)

def Cal_Min(df):
    x = df[['Experience_Years', 'Age', 'Salary']].min()
    print(x)

def Menu(df):
    print('1:For Mean')
    print('2:For Percentile')
    print('3:For Median')
    print('4:For Max')
    print('5:For Min')
    while True:
        print(" ")
        a = int(input('Enter your choice: '))
        if a == 1:
            Cal_Mean(df)
        elif a == 2:
            Cal_Perc(df)
        elif a == 3:
            Cal_Median(df)
        elif a == 4:
            Cal_Max(df)
        elif a == 5:
            Cal_Min(df)
        else:
            print('Invalid choice')
        b = input('Do you want to continue?(Y/N)')
        if b == 'N':
            break

Menu(df)
```
